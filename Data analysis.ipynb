```
# Most Common Accident Factors by City

# Get the list of cities from the filtered DataFrame
cities_list = df_cities['District'].unique()

print("Most Common Accident Factors by City:")

# List to store the most common factor and count for each city
most_common_factors_list = []

for city in cities_list:
    # Filter the DataFrame for the current city
    df_city = df_cities[df_cities['District'] == city]

    # Group by Factors and count the occurrences for the current city
    factor_counts_city = df_city.groupby('Factors').size().reset_index(name='count')

    # Find the index of the maximum count for the current city
    if not factor_counts_city.empty:
        most_common_factor_city = factor_counts_city.loc[factor_counts_city['count'].idxmax()]

        print(f"\nCity: {city}")
        print(f"  Most Common Factor: {most_common_factor_city['Factors']}")
        print(f"  Count: {most_common_factor_city['count']}")

        # Append the results to the list
        most_common_factors_list.append({
            'District': city,
            'Factors': most_common_factor_city['Factors'],
            'count': most_common_factor_city['count']
        })
    else:
        print(f"\nCity: {city}")
        print("  No data available for this city.")

# Convert the list of dictionaries to a DataFrame for plotting
most_common_factors = pd.DataFrame(most_common_factors_list)

# Now visualize it using seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=most_common_factors, x='District', y='count', hue='Factors', dodge=False)
plt.title('Most Common Accident Factors by District', fontsize=16, fontweight='bold')
plt.xlabel('District', fontsize=12)
plt.ylabel('Count of Most Common Factor', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout()
plt.show()
```

## Prepare text data

### Subtask:
Extract the text from the 'Factors' column and combine it into a single string.


**Reasoning**:
Extract the 'Factors' column, convert it to a list, and join the elements into a single string.



```
text = ' '.join(df['Factors'].tolist())
```

## Generate word cloud

### Subtask:
Create a WordCloud object and generate the word cloud from the text data.


**Reasoning**:
Instantiate a WordCloud object and generate the word cloud from the text data.



```
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white', random_state=42).generate(text)
```

**Reasoning**:
The word cloud has been generated, now visualize it.



```
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
